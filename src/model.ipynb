```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data = pd.read_csv('../power_data/AEP_hourly.csv', header=None, names=['timestamp', 'consumption'])
```

```python
data['hour'] = data['timestamp'].str.split(' ', expand=True)[1].str.split(':', expand=True)[0].astype(int)
data['consumption'] = data['consumption'].astype(float)
X = data['hour'].values
y = data['consumption'].values
```

```python
# Create a line plot
plt.plot(X, y)

# Customize the plot (optional)
plt.title('Hourly Consumption')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot (or save it to a file using plt.savefig)
plt.show()
```
